```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

merged = gpd.GeoDataFrame.merge(df_index, left_on='community', right_on='COMMUNITY')

# 4. Project to a "flat" coordinate system for Chicago (EPSG:3435)
# This ensures Chicago looks like its actual shape rather than tilted
merged = merged.to_crs(epsg=3435)

# 5. Plot the heatmap
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
merged.plot(column='transit_index_score', 
            ax=ax, 
            cmap='viridis', 
            legend=True, 
            edgecolor='black', 
            linewidth=0.2)

ax.set_title("Chicago Transit Accessibility Index", fontsize=16)
ax.axis('off')
plt.show()
```
