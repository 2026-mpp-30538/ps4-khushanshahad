```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

def scrape_oig_marathon():
    base_url = "https://oig.hhs.gov/fraud/enforcement/"
    session = requests.Session()
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    })
    
    all_data = []
    current_page = 1
    
    print("🚀 Starting marathon scrape of 500+ pages...")

    try:
        while True:
            # Construct URL for the specific page
            url = f"{base_url}?page={current_page}"
            response = session.get(url, timeout=20)
            
            if response.status_code != 200:
                print(f"\n❌ Server returned status {response.status_code} at page {current_page}. Stopping.")
                break
            
            soup = BeautifulSoup(response.text, 'html.parser')
            cards = soup.find_all('li', class_='usa-card')
            
            # If we find no cards, we are definitely at the end
            if not cards:
                print("\n🏁 No more content found. Scraping complete.")
                break
                
            for card in cards:
                # 1. Heading (Title and Link)
                h3 = card.find('h3', class_='usa-card__heading')
                link_tag = h3.find('a') if h3 else None
                
                title = link_tag.get_text(strip=True) if link_tag else "N/A"
                link = "https://oig.hhs.gov" + link_tag['href'] if link_tag and link_tag['href'].startswith('/') else "N/A"
                
                # 2. Bottom Left (Date)
                date_tag = card.find('span', class_='text-base-dark')
                date = date_tag.get_text(strip=True) if date_tag else "N/A"
                
                # 3. Bottom Right (Category)
                category_tag = card.find('ul', class_='usa-card__media-description')
                category = category_tag.get_text(strip=True) if category_tag else "N/A"
                
                all_data.append({
                    "Date": date,
                    "Title": title,
                    "Category": category,
                    "Link": link
                })
            
            print(f"✅ Scraped Page {current_page} | Total Items: {len(all_data)}", end="\r")

            # SAVE CHECKPOINT: Every 10 pages, save to CSV so you don't lose progress
            if current_page % 10 == 0:
                pd.DataFrame(all_data).to_csv("oig_scraping_checkpoint.csv", index=False)

            # PAGINATION CHECK: Look for the 'Next' link
            # We look for the 'a' tag that has the 'aria-label' containing "Next page"
            next_page = soup.find('a', attrs={'aria-label': 'Next page'})
            
            if next_page:
                current_page += 1
                # Small sleep to be respectful to the server
                time.sleep(0.5) 
            else:
                print("\n✅ Reached the last page successfully.")
                break

    except KeyboardInterrupt:
        print("\n🛑 Scrape paused by user. Saving what we have...")
    except Exception as e:
        print(f"\n⚠️ An unexpected error occurred: {e}")

    # Final Save
    df = pd.DataFrame(all_data)
    df.to_csv("oig_enforcement_final.csv", index=False)
    return df

# Start the engine
df_final = scrape_oig_marathon()
print(f"Total entries saved: {len(df_final)}")
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
from datetime import datetime

def scrape_oig_actions(start_month, start_year, run_scraper=False):
    # 1. Validation & Toggle
    if start_year < 2013:
        print("Please restrict to year >= 2013.")
        return None
    if not run_scraper:
        print("Indicator off. Skipping scrape.")
        return None

    # Threshold date
    start_threshold = datetime(start_year, start_month, 1)
    
    base_url = "https://oig.hhs.gov/fraud/enforcement/"
    session = requests.Session()
    session.headers.update({"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"})
    
    all_data = []
    current_page = 1
    keep_going = True

    while keep_going:
        url = f"{base_url}?page={current_page}"
        print(f"Scraping page {current_page}...", end="\r")
        
        response = session.get(url, timeout=20)
        if response.status_code != 200:
            break
            
        soup = BeautifulSoup(response.text, 'html.parser')
        cards = soup.find_all('li', class_='usa-card')
        
        if not cards:
            break
            
        for card in cards:
            # 1. Extract Date
            date_tag = card.find('span', class_='text-base-dark')
            if not date_tag: continue
            
            # Parse: "February 3, 2026"
            try:
                item_date = datetime.strptime(date_tag.get_text(strip=True), '%B %d, %Y')
            except ValueError:
                continue

            # STOP logic: If item is older than threshold
            if item_date < start_threshold:
                keep_going = False
                break

            # 2. Extract Title and Link (Updated to h2)
            h2_tag = card.find('h2', class_='usa-card__heading')
            link_tag = h2_tag.find('a') if h2_tag else None
            title = link_tag.get_text(strip=True) if link_tag else "N/A"
            link = "https://oig.hhs.gov" + link_tag['href'] if link_tag else "N/A"

            # 3. Extract Category (Updated to usa-tag)
            # Sometimes there are multiple tags; we'll join them just in case
            tags = card.find_all('li', class_='usa-tag')
            category = ", ".join([t.get_text(strip=True) for t in tags]) if tags else "Criminal and Civil Actions"

            all_data.append({
                "Date": item_date.strftime('%Y-%m-%d'),
                "Title": title,
                "Category": category,
                "Link": link
            })

        # Check for pagination-next button
        if not soup.find('a', class_='pagination-next') or not keep_going:
            break
            
        current_page += 1
        time.sleep(1) # PSET Requirement: 1s sleep

    # Save to CSV
    df = pd.DataFrame(all_data)
    filename = f"enforcement_actions_{start_year}_{start_month}.csv"
    df.to_csv(filename, index=False)
    print(f"\n✅ Done! Saved {len(df)} actions to {filename}")
    return df

# --- RUNNING THE TASKS ---
# Step B: Since Jan 2024
# df_2024 = scrape_oig_actions(1, 2024, run_scraper=True)

# Step C: Since Jan 2022 (The main dataset for Step 3)
df_main = scrape_oig_actions(1, 2022, run_scraper=True)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
from datetime import datetime

def scrape_oig_marathon(start_month, start_year, run_scraper=False):
    # 1. Year Validation
    if start_year < 2013:
        print("Please restrict search to year >= 2013.")
        return None
    
    # 2. Execution Indicator
    filename = f"enforcement_actions_{start_year}_{start_month}.csv"
    if not run_scraper:
        print(f"run_scraper=False. Loading {filename} if available.")
        try:
            return pd.read_csv(filename)
        except FileNotFoundError:
            print("CSV not found. Please set run_scraper=True once.")
            return None

    # Threshold Date
    start_threshold = datetime(start_year, start_month, 1)
    
    base_url = "https://oig.hhs.gov/fraud/enforcement/"
    session = requests.Session()
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"
    })
    
    all_data = []
    current_page = 1
    keep_going = True

    print(f"🚀 Scraping from {start_month}/{start_year} back to start...")

    while keep_going:
        url = f"{base_url}?page={current_page}"
        try:
            response = session.get(url, timeout=20)
            if response.status_code != 200:
                print(f"\nStopped at page {current_page} (Status {response.status_code})")
                break
            
            soup = BeautifulSoup(response.text, 'html.parser')
            cards = soup.find_all('li', class_='usa-card')
            
            # If a page is empty, we've gone too far
            if not cards:
                print(f"\nNo more cards found at page {current_page}.")
                break
                
            for card in cards:
                # 1. Extract and Parse Date
                date_tag = card.find('span', class_='text-base-dark')
                if not date_tag: continue
                
                try:
                    item_date = datetime.strptime(date_tag.get_text(strip=True), '%B %d, %Y')
                except ValueError:
                    continue

                # EXIT CONDITION: Check if we have passed the user's requested date
                if item_date < start_threshold:
                    keep_going = False
                    break

                # 2. Extract Title and Link (h2 based on your provided HTML)
                h2_tag = card.find('h2', class_='usa-card__heading')
                link_tag = h2_tag.find('a') if h2_tag else None
                title = link_tag.get_text(strip=True) if link_tag else "N/A"
                link = "https://oig.hhs.gov" + link_tag['href'] if link_tag else "N/A"

                # 3. Extract Category (all usa-tag list items)
                tags = card.find_all('li', class_='usa-tag')
                category = ", ".join([t.get_text(strip=True) for t in tags]) if tags else "N/A"

                all_data.append({
                    "Date": item_date.strftime('%Y-%m-%d'),
                    "Title": title,
                    "Category": category,
                    "Link": link
                })

            print(f"✅ Scraped Page {current_page} | Collected: {len(all_data)} items", end="\r")
            
            # Increment and Sleep
            current_page += 1
            time.sleep(1) # PSET Requirement: 1s sleep per page

        except Exception as e:
            print(f"\nError encountered on page {current_page}: {e}")
            break

    # Save to DataFrame and CSV
    df = pd.DataFrame(all_data)
    df.to_csv(filename, index=False)
    print(f"\n🏁 Complete. Saved {len(df)} rows to {filename}")
    return df

# --- TASK B: Since Jan 2024 ---
# df_2024 = scrape_oig_marathon(1, 2024, run_scraper=True)

# --- TASK C: Since Jan 2022 ---
df_final = scrape_oig_marathon(1, 2022, run_scraper=True)
```
